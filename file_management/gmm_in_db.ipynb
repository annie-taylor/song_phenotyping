```python
import pandas as pd

def filter_csv_by_identifiers(file_path, identifier_list):
    """
    Read CSV file and return subset matching the identifier list

    Args:
        file_path (str): Path to the CSV file
        identifier_list (list): List of string identifiers to filter by

    Returns:
        pandas.DataFrame: Filtered subset of the CSV
    """
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Filter rows where first column values are in the identifier list
    # Assuming first column contains the string identifiers
    first_col = df.columns[0]
    filtered_df = df[df[first_col].isin(identifier_list)]

    return filtered_df
```
