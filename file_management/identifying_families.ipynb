```python
def create_tutoring_cohorts(birds_with_dates, learning_period=90, min_overlap=15):
    """
    Create tutoring cohorts based on overlapping learning periods

    Args:
        birds_with_dates: DataFrame with birds that have valid birth dates
        learning_period: Length of learning period in days (default 90)
        min_overlap: Minimum overlap in days to be in same cohort (default 15)

    Returns:
        list: Cohort information for each bird
    """
    from datetime import timedelta

    # Add learning period dates
    birds_with_dates['Learning Start'] = birds_with_dates['Birth Date Parsed']
    birds_with_dates['Learning End'] = birds_with_dates['Birth Date Parsed'] + timedelta(days=learning_period)

    cohort_data = []

    # Group by nest environment
    nest_groups = birds_with_dates.groupby(['Nest Father', 'Nest Mother'])

    for (nest_father, nest_mother), group in nest_groups:
        if nest_father == 'Unknown' or nest_mother == 'Unknown':
            # Handle unknown nest parents separately
            for _, bird in group.iterrows():
                cohort_data.append({
                    'Bird Name': bird['Bird Name'],
                    'Rearing Type': bird['Rearing Type'],
                    'Birth Date': bird['Birth Date Parsed'].strftime('%Y-%m-%d'),
                    'Nest Father': nest_father,
                    'Nest Mother': nest_mother,
                    'Genetic Father': bird['Genetic Father'],  # ADD THIS LINE
                    'Genetic Mother': bird['Genetic Mother'],  # ADD THIS LINE
                    'Cohort ID': 'Unknown_Nest',
                    'Cohort Size': 1,
                    'Cohort Learning Period': f"{bird['Learning Start'].strftime('%Y-%m-%d')} to {bird['Learning End'].strftime('%Y-%m-%d')}",
                    'Cohort Mates': 'None',
                    'Total Songs': bird['Total Songs']
                })
            continue

        # Find overlapping learning periods within this nest group
        birds_list = group.to_dict('records')
        cohorts = []

        for bird in birds_list:
            # Find which existing cohort this bird belongs to
            assigned = False
            for cohort in cohorts:
                # Check if bird's learning period overlaps with any bird in this cohort
                for cohort_bird in cohort:
                    overlap_start = max(bird['Learning Start'], cohort_bird['Learning Start'])
                    overlap_end = min(bird['Learning End'], cohort_bird['Learning End'])
                    overlap_days = (overlap_end - overlap_start).days

                    if overlap_days >= min_overlap:
                        cohort.append(bird)
                        assigned = True
                        break
                if assigned:
                    break

            # If not assigned to existing cohort, create new one
            if not assigned:
                cohorts.append([bird])

        # Process each cohort
        for cohort_num, cohort_birds in enumerate(cohorts, 1):
            cohort_id = f"{nest_father}_{nest_mother}_cohort_{cohort_num}"

            # Calculate cohort learning period (earliest start to latest end)
            earliest_start = min(bird['Learning Start'] for bird in cohort_birds)
            latest_end = max(bird['Learning End'] for bird in cohort_birds)
            cohort_learning_period = f"{earliest_start.strftime('%Y-%m-%d')} to {latest_end.strftime('%Y-%m-%d')}"

            # Create cohort mates list
            all_bird_names = [bird['Bird Name'] for bird in cohort_birds]

            for bird in cohort_birds:
                cohort_mates = [name for name in all_bird_names if name != bird['Bird Name']]

                cohort_data.append({
                    'Bird Name': bird['Bird Name'],
                    'Rearing Type': bird['Rearing Type'],
                    'Birth Date': bird['Birth Date Parsed'].strftime('%Y-%m-%d'),
                    'Nest Father': nest_father,
                    'Nest Mother': nest_mother,
                    'Genetic Father': bird['Genetic Father'],
                    'Genetic Mother': bird['Genetic Mother'],
                    'Cohort ID': cohort_id,
                    'Cohort Size': len(cohort_birds),
                    'Cohort Learning Period': cohort_learning_period,
                    'Cohort Mates': '; '.join(cohort_mates) if cohort_mates else 'None',
                    'Total Songs': bird['Total Songs']
                })

    return cohort_data
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from collections import defaultdict
import os

def parse_birth_date(date_str):
    """
    Parse birth date from various formats, return datetime object or None
    """
    if pd.isna(date_str) or date_str == 'Unknown' or date_str == '':
        return None

    # Try different date formats
    date_formats = [
        '%Y-%m-%d',
        '%m/%d/%Y',
        '%d/%m/%Y',
        '%Y/%m/%d',
        '%m-%d-%Y',
        '%d-%m-%Y',
        '%Y%m%d'
    ]

    date_str = str(date_str).strip()

    for fmt in date_formats:
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            continue

    return None

def create_clutch_groups(father_summary_df, valid_male_xfosters, home_reared_subset,
                        clutch_window=30, output_dir='.'):
    """
    Create clutch groups based on nest environment and birth dates

    Args:
        father_summary_df: DataFrame with father summary data
        valid_male_xfosters: DataFrame with cross-foster birds
        home_reared_subset: DataFrame with home-reared birds
        clutch_window: Days within which birds are considered same clutch
        output_dir: Directory to save CSV files
    """

    # Combine all birds with rearing type
    all_birds_list = []

    # Add cross-foster birds
    for _, bird in valid_male_xfosters.iterrows():
        all_birds_list.append({
            'Bird Name': bird['Bird Name'],
            'Rearing Type': 'xf',
            'Birth Date': bird.get('DOB', 'Unknown'),
            'Nest Father': bird.get('Nest Father', 'Unknown'),
            'Nest Mother': bird.get('Nest Mother', 'Unknown'),
            'Genetic Father': bird.get('Genetic Father', 'Unknown'),
            'Genetic Mother': bird.get('Genetic Mother', 'Unknown'),
            'Total Songs': bird.get('Total Songs', 0)
        })

    # Add home-reared birds
    for _, bird in home_reared_subset.iterrows():
        all_birds_list.append({
            'Bird Name': bird['Bird Name'],
            'Rearing Type': 'hr',
            'Birth Date': bird.get('DOB', 'Unknown'),
            'Nest Father': bird.get('Nest Father', 'Unknown'),
            'Nest Mother': bird.get('Nest Mother', 'Unknown'),
            'Genetic Father': bird.get('Genetic Father', 'Unknown'),
            'Genetic Mother': bird.get('Genetic Mother', 'Unknown'),
            'Total Songs': bird.get('Total Songs', 0)
        })

    all_birds_df = pd.DataFrame(all_birds_list)

    # Parse birth dates
    all_birds_df['Birth Date Parsed'] = all_birds_df['Birth Date'].apply(parse_birth_date)

    # Separate birds with and without valid birth dates
    birds_with_dates = all_birds_df[all_birds_df['Birth Date Parsed'].notna()].copy()
    missing_birthdate_birds = all_birds_df[all_birds_df['Birth Date Parsed'].isna()].copy()

    print(f"Birds with valid birth dates: {len(birds_with_dates)}")
    print(f"Birds with missing birth dates: {len(missing_birthdate_birds)}")

    # Group by nest environment (nest father + nest mother)
    clutch_groups = {}
    detailed_birds = []
    clutch_summaries = []

    # Group birds by nest parents
    nest_groups = birds_with_dates.groupby(['Nest Father', 'Nest Mother'])

    for (nest_father, nest_mother), group in nest_groups:
        if nest_father == 'Unknown' or nest_mother == 'Unknown':
            # Handle birds with unknown nest parents separately
            continue

        # Sort by birth date
        group_sorted = group.sort_values('Birth Date Parsed')

        # Create clutches using sliding window approach
        clutches = []
        current_clutch = []

        for _, bird in group_sorted.iterrows():
            if not current_clutch:
                # Start new clutch
                current_clutch = [bird]
            else:
                # Check if bird fits in current clutch (within clutch_window days)
                days_diff = (bird['Birth Date Parsed'] - current_clutch[0]['Birth Date Parsed']).days

                if days_diff <= clutch_window:
                    current_clutch.append(bird)
                else:
                    # Close current clutch and start new one
                    clutches.append(current_clutch)
                    current_clutch = [bird]

        # Don't forget the last clutch
        if current_clutch:
            clutches.append(current_clutch)

        # Process each clutch
        for clutch_num, clutch_birds in enumerate(clutches, 1):
            clutch_id = f"{nest_father}_{nest_mother}_{clutch_num}"

            # Get clutch info
            birth_dates = [bird['Birth Date Parsed'] for bird in clutch_birds]
            min_date = min(birth_dates)
            max_date = max(birth_dates)
            date_range = f"{min_date.strftime('%Y-%m-%d')} to {max_date.strftime('%Y-%m-%d')}"

            # Separate by rearing type
            hr_birds = [bird['Bird Name'] for bird in clutch_birds if bird['Rearing Type'] == 'hr']
            xf_birds = [bird['Bird Name'] for bird in clutch_birds if bird['Rearing Type'] == 'xf']
            gen_fathers = [bird['Genetic Father'] for bird in clutch_birds]
            all_bird_names = [bird['Bird Name'] for bird in clutch_birds]

            # Create clutch mates list (excluding self)
            for bird in clutch_birds:
                clutch_mates = [name for name in all_bird_names if name != bird['Bird Name']]

                detailed_birds.append({
                    'Bird Name': bird['Bird Name'],
                    'Rearing Type': bird['Rearing Type'],
                    'Birth Date': bird['Birth Date Parsed'].strftime('%Y-%m-%d'),
                    'Nest Father': nest_father,
                    'Nest Mother': nest_mother,
                    'Genetic Father': bird['Genetic Father'],
                    'Genetic Mother': bird['Genetic Mother'],
                    'Clutch ID': clutch_id,
                    'Clutch Size': len(clutch_birds),
                    'Date Range': date_range,
                    'Clutch Mates': '; '.join(clutch_mates) if clutch_mates else 'None',
                    'Total Songs': bird['Total Songs']
                })

            # Create summary for this clutch
            birth_dates_str = '; '.join([f"{bird['Bird Name']}:{bird['Birth Date Parsed'].strftime('%Y-%m-%d')}"
                                       for bird in clutch_birds])

            clutch_summaries.append({
                'Clutch ID': clutch_id,
                'Nest Father': nest_father,
                'Nest Mother': nest_mother,
                'Clutch Size': len(clutch_birds),
                'HR Count': len(hr_birds),
                'XF Count': len(xf_birds),
                'Date Range': date_range,
                'HR Birds': '; '.join(hr_birds) if hr_birds else 'None',
                'XF Birds': '; '.join(xf_birds) if xf_birds else 'None',
                'Genetic Fathers': '; '.join(gen_fathers) if gen_fathers else 'None',
                'All Birth Dates': birth_dates_str
            })

    # Handle birds with unknown nest parents
    unknown_nest_birds = birds_with_dates[
        (birds_with_dates['Nest Father'] == 'Unknown') |
        (birds_with_dates['Nest Mother'] == 'Unknown')
    ]

    for _, bird in unknown_nest_birds.iterrows():
        bird.keys()
        detailed_birds.append({
            'Bird Name': bird['Bird Name'],
            'Rearing Type': bird['Rearing Type'],
            'Birth Date': bird['Birth Date Parsed'].strftime('%Y-%m-%d'),
            'Nest Father': bird['Nest Father'],
            'Nest Mother': bird['Nest Mother'],
            'Genetic Father': bird['Genetic Father'],
            'Genetic Mother': bird['Genetic Mother'],
            'Clutch ID': 'Unknown_Nest',
            'Clutch Size': 1,
            'Date Range': bird['Birth Date Parsed'].strftime('%Y-%m-%d'),
            'Clutch Mates': 'None',
            'Total Songs': bird['Total Songs']
        })

    # Create DataFrames
    detailed_df = pd.DataFrame(detailed_birds)
    summary_df = pd.DataFrame(clutch_summaries)

    # Prepare missing birthdate birds with reason
    missing_df = missing_birthdate_birds.copy()
    missing_df['Reason'] = missing_df['Birth Date'].apply(
        lambda x: 'Missing birth date' if pd.isna(x) or x == 'Unknown' or x == ''
        else 'Invalid date format'
    )
    missing_df = missing_df[['Bird Name', 'Rearing Type', 'Nest Father', 'Nest Mother', 'Genetic Father', 'Genetic Mother', 'Birth Date', 'Reason', 'Total Songs']]

    # CREATE TUTORING COHORTS
    logging.info("Creating tutoring cohorts based on overlapping learning periods...")
    cohort_data = create_tutoring_cohorts(birds_with_dates, learning_period=90, min_overlap=15)
    cohort_df = pd.DataFrame(cohort_data)

    # Create cohort summary
    cohort_summary = []
    cohort_groups = cohort_df.groupby('Cohort ID')

    for cohort_id, group in cohort_groups:
        if cohort_id == 'Unknown_Nest':
            continue

        hr_birds = group[group['Rearing Type'] == 'hr']['Bird Name'].tolist()
        xf_birds = group[group['Rearing Type'] == 'xf']['Bird Name'].tolist()
        print(group.keys())
        gen_faths = group['Genetic Father'].tolist()

        cohort_summary.append({
            'Cohort ID': cohort_id,
            'Nest Father': group.iloc[0]['Nest Father'],
            'Nest Mother': group.iloc[0]['Nest Mother'],
            'Cohort Size': len(group),
            'HR Count': len(hr_birds),
            'XF Count': len(xf_birds),
            'Learning Period': group.iloc[0]['Cohort Learning Period'],
            'HR Birds': '; '.join(hr_birds) if hr_birds else 'None',
            'XF Birds': '; '.join(xf_birds) if xf_birds else 'None',
            'Genetic Fathers': '; '.join(gen_faths) if gen_faths else 'None',
        })

    cohort_summary_df = pd.DataFrame(cohort_summary)

    # Save CSV files
    detailed_file = os.path.join(output_dir, 'clutch_birds_detailed.csv')
    summary_file = os.path.join(output_dir, 'clutch_summary.csv')
    missing_file = os.path.join(output_dir, 'missing_birthdate_birds.csv')

    detailed_df.to_csv(detailed_file, index=False)
    summary_df.to_csv(summary_file, index=False)
    missing_df.to_csv(missing_file, index=False)

    print(f"\nCSV files created:")
    print(f"  {detailed_file} - {len(detailed_df)} birds with clutch assignments")
    print(f"  {summary_file} - {len(summary_df)} clutch groups")
    print(f"  {missing_file} - {len(missing_df)} birds with missing birth dates")

    # Save cohort CSV files
    cohort_detailed_file = os.path.join(output_dir, 'tutoring_cohorts_detailed.csv')
    cohort_summary_file = os.path.join(output_dir, 'tutoring_cohorts_summary.csv')

    cohort_df.to_csv(cohort_detailed_file, index=False)
    cohort_summary_df.to_csv(cohort_summary_file, index=False)

    print(f"  {cohort_detailed_file} - {len(cohort_df)} birds with cohort assignments")
    print(f"  {cohort_summary_file} - {len(cohort_summary_df)} tutoring cohorts")

    # Print some summary statistics
    print(f"\nClutch Statistics:")
    if len(summary_df) > 0:
        print(f"  Total clutches identified: {len(summary_df)}")
        print(f"  Average clutch size: {summary_df['Clutch Size'].mean():.1f}")
        print(f"  Largest clutch: {summary_df['Clutch Size'].max()} birds")
        print(f"  Clutches with both HR and XF birds: {len(summary_df[(summary_df['HR Count'] > 0) & (summary_df['XF Count'] > 0)])}")

    # Show example clutches
    print(f"\nExample clutches (first 5):")
    for _, clutch in summary_df.head().iterrows():
        print(f"  {clutch['Clutch ID']}: {clutch['Clutch Size']} birds ({clutch['HR Count']} HR, {clutch['XF Count']} XF)")
        print(f"    Date range: {clutch['Date Range']}")

    return detailed_df, summary_df, missing_df, cohort_df, cohort_summary_df

# Run the clutch analysis
print("Creating clutch groups...")
detailed_df, summary_df, missing_df, cohort_df, cohort_summary_df = create_clutch_groups(
    father_summary_df=father_summary_df,
    valid_male_xfosters=valid_male_xfosters,
    home_reared_subset=home_reared_subset,
    clutch_window=30,  # Adjust this parameter as needed
    output_dir='.'  # Current directory, change as needed
)

# Display some results for inspection
print("\nFirst few rows of detailed data:")
print(detailed_df.head())

print("\nClutch summary:")
print(summary_df.head())

print("\nMissing birthdate birds:")
print(missing_df.head())
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

def plot_learning_periods_by_father(cohort_df, nest_father):
    """Show learning periods as horizontal bars for one father's offspring"""
    father_birds = cohort_df[cohort_df['Nest Father'] == nest_father]

    fig, ax = plt.subplots(figsize=(15, 8))

    colors = {'hr': 'blue', 'xf': 'red'}

    for i, (_, bird) in enumerate(father_birds.iterrows()):
        birth_date = datetime.strptime(bird['Birth Date'], '%Y-%m-%d')
        learning_end = birth_date + timedelta(days=90)

        ax.barh(i, 90, left=birth_date,
               color=colors[bird['Rearing Type']], alpha=0.7,
               label=bird['Rearing Type'] if i == 0 else "")

        # Add bird name
        ax.text(birth_date + timedelta(days=45), i, bird['Bird Name'],
               ha='center', va='center', fontsize=8)

    ax.set_yticks(range(len(father_birds)))
    ax.set_yticklabels([f"Cohort {row['Cohort ID'].split('_')[-1]}" for _, row in father_birds.iterrows()])
    ax.set_xlabel('Timeline')
    ax.set_title(f'Learning Periods for {nest_father} Offspring')

    plt.tight_layout()
    return fig
```

```python
import seaborn as sns
import numpy as np

def create_overlap_heatmap(cohort_summary_df):
    """Heatmap showing temporal overlap between cohorts"""

    # Calculate overlap matrix
    cohorts = cohort_summary_df['Cohort ID'].tolist()
    overlap_matrix = np.zeros((len(cohorts), len(cohorts)))

    for i, cohort1 in enumerate(cohorts):
        period1 = cohort_summary_df.iloc[i]['Learning Period']
        start1, end1 = [datetime.strptime(d.strip(), '%Y-%m-%d')
                       for d in period1.split(' to ')]

        for j, cohort2 in enumerate(cohorts):
            period2 = cohort_summary_df.iloc[j]['Learning Period']
            start2, end2 = [datetime.strptime(d.strip(), '%Y-%m-%d')
                           for d in period2.split(' to ')]

            # Calculate overlap
            overlap_start = max(start1, start2)
            overlap_end = min(end1, end2)
            overlap_days = max(0, (overlap_end - overlap_start).days)

            overlap_matrix[i, j] = overlap_days

    # Create heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(overlap_matrix,
                xticklabels=[c.split('_')[-1] for c in cohorts],
                yticklabels=[c.split('_')[-1] for c in cohorts],
                annot=True, fmt='d', cmap='YlOrRd')
    plt.title('Learning Period Overlap Between Cohorts (Days)')
    plt.tight_layout()
```

```python
def create_father_timeline_grid(cohort_df):
    """Small multiples showing timeline for each father"""

    fathers = cohort_df['Nest Father'].unique()
    n_fathers = len(fathers)
    cols = 3
    rows = (n_fathers + cols - 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=(20, 5*rows))
    axes = axes.flatten() if n_fathers > 1 else [axes]

    for i, father in enumerate(fathers):
        if i >= len(axes):
            break

        father_data = cohort_df[cohort_df['Nest Father'] == father]
        ax = axes[i]

        # Plot each bird's learning period
        colors = {'hr': 'skyblue', 'xf': 'salmon'}

        for j, (_, bird) in enumerate(father_data.iterrows()):
            birth_date = datetime.strptime(bird['Birth Date'], '%Y-%m-%d')
            learning_end = birth_date + timedelta(days=90)

            ax.barh(j, 90, left=birth_date,
                   color=colors[bird['Rearing Type']], alpha=0.8)

            # Add cohort grouping
            cohort_num = bird['Cohort ID'].split('_')[-1]
            ax.text(birth_date - timedelta(days=10), j, f"C{cohort_num}",
                   ha='right', va='center', fontsize=8, weight='bold')

        ax.set_title(f'{father} ({len(father_data)} offspring)')
        ax.set_yticks(range(len(father_data)))
        ax.set_yticklabels([row['Bird Name'] for _, row in father_data.iterrows()],
                          fontsize=8)

        # Format x-axis
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=6))
        plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)

    # Hide empty subplots
    for i in range(len(fathers), len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    return fig
```

```python
def create_cohort_dashboard(cohort_df, cohort_summary_df):
    """Multi-panel dashboard with key metrics"""

    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=['Cohort Size Distribution', 'HR vs XF by Cohort',
                       'Timeline Span by Father', 'Learning Period Overlap'],
        specs=[[{"type": "bar"}, {"type": "bar"}],
               [{"type": "scatter"}, {"type": "histogram"}]]
    )

    # Panel 1: Cohort size distribution
    cohort_sizes = cohort_summary_df['Cohort Size'].value_counts().sort_index()
    fig.add_trace(
        go.Bar(x=cohort_sizes.index, y=cohort_sizes.values, name="Cohort Size"),
        row=1, col=1
    )

    # Panel 2: HR vs XF breakdown
    fig.add_trace(
        go.Bar(x=cohort_summary_df['Cohort ID'], y=cohort_summary_df['HR Count'],
               name="Home-Reared", marker_color='skyblue'),
        row=1, col=2
    )
    fig.add_trace(
        go.Bar(x=cohort_summary_df['Cohort ID'], y=cohort_summary_df['XF Count'],
               name="Cross-Foster", marker_color='salmon'),
        row=1, col=2
    )

    # Panel 3: Timeline span by father (scatter plot)
    father_stats = []
    for father in cohort_df['Nest Father'].unique():
        father_data = cohort_df[cohort_df['Nest Father'] == father]
        if len(father_data) > 1:
            dates = [datetime.strptime(d, '%Y-%m-%d') for d in father_data['Birth Date']]
            span_days = (max(dates) - min(dates)).days
            father_stats.append({'Father': father, 'Span': span_days, 'Count': len(father_data)})

    father_df = pd.DataFrame(father_stats)
    if len(father_df) > 0:
        fig.add_trace(
            go.Scatter(x=father_df['Count'], y=father_df['Span'],
                      mode='markers', text=father_df['Father'],
                      marker=dict(size=10), name="Father Timeline"),
            row=2, col=1
        )

    # Panel 4: Learning period overlap histogram
    overlaps = []
    for _, cohort in cohort_summary_df.iterrows():
        period = cohort['Learning Period']
        start, end = [datetime.strptime(d.strip(), '%Y-%m-%d')
                     for d in period.split(' to ')]
        overlap_days = (end - start).days
        overlaps.append(overlap_days)

    fig.add_trace(
        go.Histogram(x=overlaps, nbinsx=20, name="Learning Period Length"),
        row=2, col=2
    )

    fig.update_layout(height=800, showlegend=False,
                     title_text="Cross-Foster Cohort Analysis Dashboard")

    return fig
```

```python
import networkx as nx
from matplotlib.patches import Rectangle

def create_cohort_network(cohort_summary_df):
    """Network diagram showing relationships between fathers and cohorts"""

    G = nx.Graph()

    # Add nodes for fathers and cohorts
    fathers = cohort_summary_df['Nest Father'].unique()
    cohorts = cohort_summary_df['Cohort ID'].unique()

    # Add father nodes
    for father in fathers:
        father_cohorts = cohort_summary_df[cohort_summary_df['Nest Father'] == father]
        total_offspring = father_cohorts['Cohort Size'].sum()
        G.add_node(father, node_type='father', size=total_offspring)

    # Add cohort nodes and edges
    for _, cohort in cohort_summary_df.iterrows():
        cohort_id = cohort['Cohort ID']
        father = cohort['Nest Father']

        G.add_node(cohort_id, node_type='cohort',
                  hr_count=cohort['HR Count'], xf_count=cohort['XF Count'])
        G.add_edge(father, cohort_id)

    # Create layout
    pos = nx.spring_layout(G, k=3, iterations=50)

    plt.figure(figsize=(15, 10))

    # Draw father nodes
    father_nodes = [n for n in G.nodes() if G.nodes[n]['node_type'] == 'father']
    nx.draw_networkx_nodes(G, pos, nodelist=father_nodes,
                          node_color='lightblue', node_size=1000, alpha=0.8)

    # Draw cohort nodes (colored by HR/XF ratio)
    cohort_nodes = [n for n in G.nodes() if G.nodes[n]['node_type'] == 'cohort']
    cohort_colors = []
    for node in cohort_nodes:
        hr = G.nodes[node]['hr_count']
        xf = G.nodes[node]['xf_count']
        ratio = hr / (hr + xf) if (hr + xf) > 0 else 0.5
        cohort_colors.append(ratio)

    nx.draw_networkx_nodes(G, pos, nodelist=cohort_nodes,
                          node_color=cohort_colors, cmap='RdYlBu',
                          node_size=500, alpha=0.8)

    # Draw edges and labels
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    nx.draw_networkx_labels(G, pos, font_size=8)

    plt.title('Father-Cohort Network\n(Blue=More HR, Red=More XF)')
    plt.axis('off')
    plt.tight_layout()
```

```python
def load_data(save_file: str) -> Dict[str, Any]:
    """Load data from JSON file."""
    if os.path.exists(save_file):
        try:
            with open(save_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            logging.info(f"Loaded existing data from {save_file}")
            return data
        except Exception as e:
            logging.error(f"Error loading data from {save_file}: {e}")
            return {}
    return {}

def create_bird_song_ages_dataset(cohort_df, audio_data_file):
    """
    Create simple dataset: bird_name, filepath, recording_date, birth_date, age_days
    """
    audio_data = load_data(audio_data_file)

    records = []
    for _, bird in cohort_df.iterrows():
        bird_name = bird['Bird Name']
        birth_date = datetime.strptime(bird['Birth Date'], '%Y-%m-%d')

        if bird_name in audio_data:
            # Get all files with datetime info
            for file_type in ['wav_files_info', 'cbin_files_info', 'audio_from_batch_info']:
                if file_type in audio_data[bird_name]:
                    for file_info in audio_data[bird_name][file_type]:
                        if file_info.get('datetime'):
                            recording_date = datetime.fromisoformat(file_info['datetime'])
                            age_days = (recording_date - birth_date).days

                            records.append({
                                'bird_name': bird_name,
                                'filepath': file_info['filepath'],
                                'recording_date': recording_date,
                                'birth_date': birth_date,
                                'age_days': age_days,
                                'file_type': file_type.replace('_info', '')
                            })

    return pd.DataFrame(records)

def get_bird_songs_by_age_range(bird_name, min_age, max_age, song_dataset):
    """
    Get filepaths for a bird's songs within specified age range
    """
    bird_songs = song_dataset[song_dataset['bird_name'] == bird_name]
    age_filtered = bird_songs[(bird_songs['age_days'] >= min_age) &
                             (bird_songs['age_days'] <= max_age)]
    return age_filtered['filepath'].tolist()
```

```python
def create_song_age_summaries(cohort_df, audio_data_file):
    """
    Create both cohort-level and father-level summaries of song age coverage
    """
    # First create the basic song dataset
    song_dataset = create_bird_song_ages_dataset(cohort_df, audio_data_file)

    # Merge with cohort info
    song_data_with_cohorts = song_dataset.merge(
        cohort_df[['Bird Name', 'Cohort ID', 'Nest Father', 'Rearing Type']],
        left_on='bird_name', right_on='Bird Name'
    )

    # Create cohort summary
    cohort_summary = create_cohort_song_summary(song_data_with_cohorts)

    # Create father summary
    father_summary = create_father_song_summary(song_data_with_cohorts)

    return cohort_summary, father_summary

def add_age_categories(song_data_with_cohorts):
    """Add age category column to the dataset"""
    def categorize_age(age_days):
        if age_days < 30:
            return '<30 days'
        elif age_days < 60:
            return '30-60 days'
        elif age_days < 90:
            return '60-90 days'
        else:
            return '90+ days'

    song_data_with_cohorts['Age Category'] = song_data_with_cohorts['age_days'].apply(categorize_age)
    return song_data_with_cohorts

def create_cohort_song_summary(song_data_with_cohorts):
    """Cohort-level summary table with age categories"""
    # Add age categories
    song_data_with_cohorts = add_age_categories(song_data_with_cohorts)

    summaries = []

    for cohort_id in song_data_with_cohorts['Cohort ID'].unique():
        cohort_songs = song_data_with_cohorts[song_data_with_cohorts['Cohort ID'] == cohort_id]

        if len(cohort_songs) == 0:
            continue

        # Basic stats
        birds_with_songs = cohort_songs['bird_name'].nunique()
        hr_birds = cohort_songs[cohort_songs['Rearing Type'] == 'hr']['bird_name'].nunique()
        xf_birds = cohort_songs[cohort_songs['Rearing Type'] == 'xf']['bird_name'].nunique()

        age_stats = cohort_songs['age_days'].agg(['min', 'max', 'count'])

        # Age category counts
        age_category_counts = cohort_songs['Age Category'].value_counts()

        summaries.append({
            'Cohort ID': cohort_id,
            'Nest Father': cohort_songs['Nest Father'].iloc[0],
            'Birds with Songs': birds_with_songs,
            'HR Birds': hr_birds,
            'XF Birds': xf_birds,
            'Has Both Types': 'Yes' if hr_birds > 0 and xf_birds > 0 else 'No',
            'Total Songs': int(age_stats['count']),
            'Min Age (days)': int(age_stats['min']),
            'Max Age (days)': int(age_stats['max']),
            'Songs <30 days': age_category_counts.get('<30 days', 0),
            'Songs 30-60 days': age_category_counts.get('30-60 days', 0),
            'Songs 60-90 days': age_category_counts.get('60-90 days', 0),
            'Songs 90+ days': age_category_counts.get('90+ days', 0),
            'Age Categories Covered': len(age_category_counts)
        })

    return pd.DataFrame(summaries).sort_values('Total Songs', ascending=False)

def create_father_song_summary(song_data_with_cohorts):
    """Father/tutor-level summary table with age categories"""
    # Add age categories
    song_data_with_cohorts = add_age_categories(song_data_with_cohorts)

    summaries = []

    for father in song_data_with_cohorts['Nest Father'].unique():
        father_songs = song_data_with_cohorts[song_data_with_cohorts['Nest Father'] == father]

        # Basic stats
        total_birds = father_songs['bird_name'].nunique()
        hr_birds = father_songs[father_songs['Rearing Type'] == 'hr']['bird_name'].nunique()
        xf_birds = father_songs[father_songs['Rearing Type'] == 'xf']['bird_name'].nunique()
        cohorts = father_songs['Cohort ID'].nunique()

        age_stats = father_songs['age_days'].agg(['min', 'max', 'count'])

        # Age category counts
        age_category_counts = father_songs['Age Category'].value_counts()

        summaries.append({
            'Nest Father': father,
            'Total Birds with Songs': total_birds,
            'HR Birds': hr_birds,
            'XF Birds': xf_birds,
            'Has Both Types': 'Yes' if hr_birds > 0 and xf_birds > 0 else 'No',
            'Number of Cohorts': cohorts,
            'Total Songs': int(age_stats['count']),
            'Min Age (days)': int(age_stats['min']),
            'Max Age (days)': int(age_stats['max']),
            'Songs <30 days': age_category_counts.get('<30 days', 0),
            'Songs 30-60 days': age_category_counts.get('30-60 days', 0),
            'Songs 60-90 days': age_category_counts.get('60-90 days', 0),
            'Songs 90+ days': age_category_counts.get('90+ days', 0),
            'Age Categories Covered': len(age_category_counts)
        })

    return pd.DataFrame(summaries).sort_values('Total Songs', ascending=False)
```

```python
import numpy as np
import scipy.signal
from scipy.ndimage import gaussian_filter1d
from typing import List, Tuple, Optional
import os
from datetime import datetime
from tools.spectrogram_configs import SpectrogramParams
from tools.song_io import read_audio_file

def rms_envelope(audio: np.ndarray, frame_length: int = 1024, hop_length: int = 512) -> np.ndarray:
    """Calculate RMS envelope of audio signal"""
    # Pad audio to ensure we don't lose samples at the end
    audio_padded = np.pad(audio, (0, frame_length), mode='constant')

    # Calculate RMS for each frame
    rms = []
    for i in range(0, len(audio_padded) - frame_length, hop_length):
        frame = audio_padded[i:i + frame_length]
        rms_val = np.sqrt(np.mean(frame**2))
        rms.append(rms_val)

    return np.array(rms)

def detect_syllables_rms(audio: np.ndarray, fs: int,
                        smooth_window_ms: float = 20,
                        threshold_percentile: float = 25,
                        min_syllable_ms: float = 12,
                        max_syllable_ms: float = 200) -> Tuple[np.ndarray, np.ndarray]:
    """
    Detect syllable onsets and offsets using RMS-based method

    Returns:
        onsets, offsets in milliseconds
    """

    # Calculate RMS envelope
    rms = rms_envelope(audio)

    # Rectify (already positive from RMS, but ensure no negatives)
    rms = np.abs(rms)

    # Gaussian smoothing (convert window from ms to samples)
    hop_length = 512  # from rms_envelope
    rms_fs = fs / hop_length  # effective sampling rate of RMS signal
    smooth_sigma = (smooth_window_ms / 1000) * rms_fs / 2.355  # convert to sigma

    smoothed = gaussian_filter1d(rms, sigma=smooth_sigma)

    # Threshold based on percentile
    threshold = np.percentile(smoothed, threshold_percentile)

    # Find crossings
    above_threshold = smoothed > threshold
    crossings = np.diff(above_threshold.astype(int))

    # Onsets where signal goes above threshold, offsets where it goes below
    onset_indices = np.where(crossings == 1)[0] + 1
    offset_indices = np.where(crossings == -1)[0] + 1

    # Handle edge cases
    if len(onset_indices) == 0 or len(offset_indices) == 0:
        return np.array([]), np.array([])

    # Ensure we start with an onset and end with an offset
    if len(offset_indices) == 0 or (len(onset_indices) > 0 and onset_indices[0] < offset_indices[0]):
        # First onset comes before first offset, good
        pass
    else:
        # Signal starts above threshold
        onset_indices = np.concatenate([[0], onset_indices])

    if len(onset_indices) > len(offset_indices):
        # Signal ends above threshold
        offset_indices = np.concatenate([offset_indices, [len(smoothed) - 1]])

    # Convert indices to milliseconds
    onsets_ms = (onset_indices * hop_length / fs) * 1000
    offsets_ms = (offset_indices * hop_length / fs) * 1000

    # Filter by syllable duration
    durations = offsets_ms - onsets_ms
    valid_mask = (durations >= min_syllable_ms) & (durations <= max_syllable_ms)

    return onsets_ms[valid_mask], offsets_ms[valid_mask]

def find_song_bouts(onsets: np.ndarray, offsets: np.ndarray,
                   min_syllables: int = 4,
                   window_ms: float = 2000) -> List[Tuple[float, float]]:
    """
    Find song bouts (continuous periods with sufficient syllable density)

    Returns:
        List of (start_ms, end_ms) tuples for each song bout
    """
    if len(onsets) < min_syllables:
        return []

    song_bouts = []

    # Sliding window approach
    for i in range(len(onsets) - min_syllables + 1):
        window_start = onsets[i]

        # Find how many syllables fit in the window
        syllables_in_window = 0
        window_end = window_start

        for j in range(i, len(onsets)):
            if onsets[j] <= window_start + window_ms:
                syllables_in_window += 1
                window_end = max(window_end, offsets[j])
            else:
                break

        # If we found enough syllables, this is a song bout
        if syllables_in_window >= min_syllables:
            bout_start = onsets[i]
            bout_end = window_end

            # Avoid overlapping bouts
            if not song_bouts or bout_start > song_bouts[-1][1]:
                song_bouts.append((bout_start, bout_end))

    return song_bouts

# Test function to validate on a single file
def test_song_detection(audio_file: str, params: SpectrogramParams):
    """Test the song detection pipeline on a single file"""

    # Read audio
    audio, fs = read_audio_file(audio_file)

    # Detect syllables
    onsets, offsets = detect_syllables_rms(audio, fs)

    # Find song bouts
    song_bouts = find_song_bouts(onsets, offsets)

    print(f"File: {os.path.basename(audio_file)}")
    print(f"Detected {len(onsets)} syllables")
    print(f"Found {len(song_bouts)} song bouts")

    for i, (start, end) in enumerate(song_bouts):
        duration = (end - start) / 1000  # convert to seconds
        print(f"  Bout {i+1}: {start:.0f}-{end:.0f}ms ({duration:.2f}s)")

    return onsets, offsets, song_bouts
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.use('MacOSX')

def visualize_detection(audio_file, onsets, offsets, song_bouts, save_path=None):
    """Visualize the RMS detection results"""
    from tools.audio_utils import read_audio_file

    # Read and process audio
    audio, fs = read_audio_file(audio_file)

    # Calculate RMS envelope (same as in detection)
    rms = rms_envelope(audio)
    hop_length = 512
    rms_fs = fs / hop_length

    # Time axis for RMS
    rms_time = np.arange(len(rms)) / rms_fs

    # Plot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

    # Raw audio
    audio_time = np.arange(len(audio)) / fs
    ax1.plot(audio_time, audio, 'b-', alpha=0.7)
    ax1.set_ylabel('Amplitude')
    ax1.set_title(f'Raw Audio: {os.path.basename(audio_file)}')

    # RMS with detections
    ax2.plot(rms_time, rms, 'k-', linewidth=1)
    ax2.set_ylabel('RMS')
    ax2.set_xlabel('Time (s)')
    ax2.set_title('RMS Envelope with Detected Syllables and Song Bouts')

    # Mark syllables
    for onset, offset in zip(onsets/1000, offsets/1000):
        ax2.axvspan(onset, offset, alpha=0.3, color='red', label='Syllables' if onset == onsets[0]/1000 else "")

    # Mark song bouts
    for i, (start, end) in enumerate(song_bouts):
        ax2.axvspan(start/1000, end/1000, alpha=0.2, color='green',
                   label='Song Bouts' if i == 0 else "")

    ax2.legend()
    plt.tight_layout()

    # Save instead of show
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        print(f"Plot saved to: {save_path}")
    else:
        plt.savefig('song_detection_test.png', dpi=150, bbox_inches='tight')
        print("Plot saved to: song_detection_test.png")

    plt.close()
def smooth_data_matlab_style(audio: np.ndarray, fs: int,
                            sm_win_ms: float = 2.0,  # 2ms smoothing window
                            f_low: float = 500.0,
                            f_high: float = 10000.0) -> np.ndarray:
    """
    Matlab-style smoothing: bandpass -> square -> moving average
    """
    from scipy.signal import butter, filtfilt

    # Bandpass filter
    nyquist = fs / 2
    low_norm = f_low / nyquist
    high_norm = f_high / nyquist

    if high_norm >= 1.0:
        high_norm = 0.99

    b, a = butter(4, [low_norm, high_norm], btype='band')
    filtsong = filtfilt(b, a, audio)

    # Square the signal (power)
    squared_song = filtsong ** 2

    # Moving average smoothing (boxcar filter)
    smooth_len = round(fs * sm_win_ms / 1000)
    h = np.ones(smooth_len) / smooth_len
    smooth = np.convolve(h, squared_song, mode='same')

    return smooth

def segment_notes_matlab_style(smooth: np.ndarray, fs: int,
                              threshold: float,
                              min_int_ms: float = 5.0,   # minimum interval between notes
                              min_dur_ms: float = 10.0) -> Tuple[np.ndarray, np.ndarray]:
    """
    Matlab-style segmentation using derivative threshold
    """

    # Derivative-based detection
    diff_smooth = np.abs(np.diff(smooth))
    notetimes = (diff_smooth > threshold).astype(int)

    # Find transitions
    h = np.array([1, -1])
    trans = np.convolve(h, notetimes, mode='valid')

    t_onsets = np.where(trans > 0)[0]
    t_offsets = np.where(trans < 0)[0]

    if len(t_onsets) == 0 or len(t_offsets) == 0:
        return np.array([]), np.array([])

    # Make sure we have matching onsets/offsets
    if len(t_onsets) != len(t_offsets):
        min_len = min(len(t_onsets), len(t_offsets))
        t_onsets = t_onsets[:min_len]
        t_offsets = t_offsets[:min_len]

    # Eliminate short intervals between notes
    if len(t_onsets) > 1:
        temp_int = (t_onsets[1:] - t_offsets[:-1]) * 1000 / fs
        real_ints = temp_int > min_int_ms

        # Keep first onset and offsets that meet interval criteria
        keep_onsets = [True] + list(real_ints)
        keep_offsets = list(real_ints) + [True]

        t_onsets = t_onsets[keep_onsets]
        t_offsets = t_offsets[keep_offsets]

    # Eliminate short notes
    temp_dur = (t_offsets - t_onsets) * 1000 / fs
    real_durs = temp_dur > min_dur_ms

    t_onsets = t_onsets[real_durs]
    t_offsets = t_offsets[real_durs]

    # Convert to seconds
    onsets = t_onsets / fs
    offsets = t_offsets / fs

    return onsets * 1000, offsets * 1000  # return in ms for consistency

def detect_syllables_matlab_style(audio: np.ndarray, fs: int,
                                 sm_win_ms: float = 2.0,
                                 threshold: float = 0.001,  # will need tuning
                                 min_int_ms: float = 5.0,
                                 min_dur_ms: float = 10.0) -> Tuple[np.ndarray, np.ndarray]:
    """
    Matlab-style syllable detection
    """

    # Smooth the data
    smooth = smooth_data_matlab_style(audio, fs, sm_win_ms)

    # Segment notes
    onsets, offsets = segment_notes_matlab_style(smooth, fs, threshold, min_int_ms, min_dur_ms)

    return onsets, offsets

# Test with the MATLAB approach
def test_matlab_detection(audio_file: str):
    from tools.audio_utils import read_audio_file

    audio, fs = read_audio_file(audio_file)

    # Try different thresholds to see what works
    thresholds = [0.0001, 0.0005, 0.001, 0.005, 0.01]

    for thresh in thresholds:
        onsets, offsets = detect_syllables_matlab_style(audio, fs, threshold=thresh)
        song_bouts = find_song_bouts(onsets, offsets)

        print(f"Threshold {thresh}: {len(onsets)} syllables, {len(song_bouts)} bouts")
        if len(song_bouts) > 0:
            for i, (start, end) in enumerate(song_bouts[:3]):  # show first 3 bouts
                print(f"  Bout {i+1}: {start:.0f}-{end:.0f}ms")

    visualize_detection(audio_file, onsets, offsets, song_bouts)

# Test the MATLAB approach
test_matlab_detection(converted_test_file)
```

```python
def segment_notes_matlab_style_fixed(smooth: np.ndarray, fs: int,
                                   threshold: float,
                                   min_int_ms: float = 5.0,
                                   min_dur_ms: float = 10.0) -> Tuple[np.ndarray, np.ndarray]:
    """
    Corrected MATLAB-style segmentation using amplitude threshold
    """

    # Amplitude thresholding (not derivative!)
    notetimes = (smooth > threshold).astype(float)

    # Find transitions using convolution
    h = np.array([1, -1])
    trans = np.convolve(h, notetimes, mode='valid')

    t_onsets = np.where(trans > 0)[0]
    t_offsets = np.where(trans < 0)[0]

    if len(t_onsets) == 0 or len(t_offsets) == 0:
        return np.array([]), np.array([])

    # Handle mismatched onsets/offsets
    if len(t_onsets) != len(t_offsets):
        min_len = min(len(t_onsets), len(t_offsets))
        t_onsets = t_onsets[:min_len]
        t_offsets = t_offsets[:min_len]

    if len(t_onsets) == 0:
        return np.array([]), np.array([])

    # Eliminate short intervals between notes
    if len(t_onsets) > 1:
        temp_int = (t_onsets[1:] - t_offsets[:-1]) * 1000 / fs
        real_ints = temp_int > min_int_ms

        # Keep first onset, then only onsets/offsets that meet interval criteria
        new_onsets = [t_onsets[0]]
        new_offsets = []

        for i, keep in enumerate(real_ints):
            if keep:
                new_onsets.append(t_onsets[i+1])
                new_offsets.append(t_offsets[i])
        new_offsets.append(t_offsets[-1])  # Keep last offset

        t_onsets = np.array(new_onsets)
        t_offsets = np.array(new_offsets)

    # Eliminate short notes
    temp_dur = (t_offsets - t_onsets) * 1000 / fs
    real_durs = temp_dur > min_dur_ms

    t_onsets = t_onsets[real_durs]
    t_offsets = t_offsets[real_durs]

    # Convert to seconds, then to ms for consistency
    onsets = t_onsets / fs * 1000
    offsets = t_offsets / fs * 1000

    return onsets, offsets

def detect_syllables_matlab_style_fixed(audio: np.ndarray, fs: int,
                                       sm_win_ms: float = 2.0,
                                       threshold: float = 0.001,
                                       min_int_ms: float = 5.0,
                                       min_dur_ms: float = 10.0) -> Tuple[np.ndarray, np.ndarray]:
    """
    Corrected MATLAB-style syllable detection with amplitude thresholding
    """

    # Smooth the data
    smooth = smooth_data_matlab_style(audio, fs, sm_win_ms)

    # Segment notes using amplitude threshold
    onsets, offsets = segment_notes_matlab_style_fixed(smooth, fs, threshold, min_int_ms, min_dur_ms)

    return onsets, offsets

# Test the corrected approach
def test_corrected_matlab_detection(audio_file: str):
    from tools.audio_utils import read_audio_file

    audio, fs = read_audio_file(audio_file)

    # Try different thresholds
    thresholds = [0.0001, 0.0005, 0.001, 0.005, 0.01]

    for thresh in thresholds:
        onsets, offsets = detect_syllables_matlab_style_fixed(audio, fs, threshold=thresh)
        song_bouts = find_song_bouts(onsets, offsets)

        print(f"Threshold {thresh}: {len(onsets)} syllables, {len(song_bouts)} bouts")
        if len(song_bouts) > 0:
            for i, (start, end) in enumerate(song_bouts[:3]):
                print(f"  Bout {i+1}: {start:.0f}-{end:.0f}ms")

# Test the corrected approach
print("Testing corrected MATLAB approach (amplitude thresholding):")
test_corrected_matlab_detection(converted_test_file)
```

```python
def detect_syllables_amplitude_threshold(audio: np.ndarray, fs: int,
                                       sm_win_ms: float = 2.0,
                                       threshold_percentile: float = 75,
                                       min_int_ms: float = 5.0,
                                       min_dur_ms: float = 10.0) -> Tuple[np.ndarray, np.ndarray]:
    """
    Simple amplitude thresholding on smoothed signal (MATLAB-style)
    """

    # Smooth the data (bandpass -> square -> moving average)
    smooth = smooth_data_matlab_style(audio, fs, sm_win_ms)

    # Threshold based on percentile of smoothed signal
    threshold = np.percentile(smooth, threshold_percentile)

    # Find where signal is above threshold
    above_threshold = smooth > threshold

    # Find transitions (onsets/offsets)
    transitions = np.diff(above_threshold.astype(int))
    onsets_idx = np.where(transitions == 1)[0] + 1
    offsets_idx = np.where(transitions == -1)[0] + 1

    # Handle edge cases
    if len(onsets_idx) == 0 or len(offsets_idx) == 0:
        return np.array([]), np.array([])

    # Ensure we start with onset and end with offset
    if above_threshold[0]:  # starts above threshold
        onsets_idx = np.concatenate([[0], onsets_idx])
    if above_threshold[-1]:  # ends above threshold
        offsets_idx = np.concatenate([offsets_idx, [len(smooth) - 1]])

    # Match onsets and offsets
    min_len = min(len(onsets_idx), len(offsets_idx))
    onsets_idx = onsets_idx[:min_len]
    offsets_idx = offsets_idx[:min_len]

    # Apply duration filter
    durations = (offsets_idx - onsets_idx) * 1000 / fs
    keep_durations = durations > min_dur_ms

    onsets_idx = onsets_idx[keep_durations]
    offsets_idx = offsets_idx[keep_durations]

    # Convert to milliseconds
    onsets_ms = onsets_idx * 1000 / fs
    offsets_ms = offsets_idx * 1000 / fs

    return onsets_ms, offsets_ms

# Simple test function
def test_amplitude_thresholds(audio_file):
    """Test different percentile thresholds"""
    from tools.audio_utils import read_audio_file

    audio, fs = read_audio_file(audio_file)

    print(f"Testing amplitude thresholds on: {os.path.basename(audio_file)}")
    print("-" * 50)

    percentiles = [60, 70, 75, 80, 85, 90]
    for p in percentiles:
        onsets, offsets = detect_syllables_amplitude_threshold(
            audio, fs, threshold_percentile=p
        )
        song_bouts = find_song_bouts(onsets, offsets)

        print(f"{p:2d}th percentile: {len(onsets):3d} syllables, {len(song_bouts)} bouts")

# Test it
test_amplitude_thresholds(converted_test_file)
```

```python
def visualize_amplitude_detection(audio_file, threshold_percentile=75, save_path=None):
    """Visualize amplitude thresholding detection"""
    from tools.audio_utils import read_audio_file

    audio, fs = read_audio_file(audio_file)

    # Get smoothed signal and detections
    smooth = smooth_data_matlab_style(audio, fs)
    onsets, offsets = detect_syllables_amplitude_threshold(
        audio, fs, threshold_percentile=threshold_percentile
    )
    song_bouts = find_song_bouts(onsets, offsets)

    # Calculate threshold value
    threshold_value = np.percentile(smooth, threshold_percentile)

    # Time axes
    audio_time = np.arange(len(audio)) / fs
    smooth_time = np.arange(len(smooth)) / fs

    # Create figure
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

    # 1. Raw audio with detections
    ax1.plot(audio_time, audio, 'b-', alpha=0.7)
    ax1.set_ylabel('Amplitude')
    ax1.set_title(f'Raw Audio: {os.path.basename(audio_file)}')
    ax1.grid(True, alpha=0.3)

    # Mark syllables on audio
    for onset, offset in zip(onsets/1000, offsets/1000):
        ax1.axvspan(onset, offset, alpha=0.3, color='red',
                   label='Syllables' if onset == onsets[0]/1000 else "")

    # Mark song bouts on audio
    for start, end in song_bouts:
        ax1.axvspan(start/1000, end/1000, alpha=0.15, color='green',
                   label='Song Bouts' if start == song_bouts[0][0] else "")

    if len(onsets) > 0:
        ax1.legend()

    # 2. Smoothed signal with threshold
    ax2.plot(smooth_time, smooth, 'k-', linewidth=1, label='Smoothed Signal')
    ax2.axhline(y=threshold_value, color='red', linestyle='--',
               label=f'{threshold_percentile}th percentile: {threshold_value:.6f}')
    ax2.set_ylabel('Smoothed Power')
    ax2.set_xlabel('Time (s)')
    ax2.set_title(f'Smoothed Signal with Amplitude Threshold')
    ax2.grid(True, alpha=0.3)
    ax2.legend()

    # Mark detections on smoothed signal
    for onset, offset in zip(onsets/1000, offsets/1000):
        ax2.axvspan(onset, offset, alpha=0.3, color='red')

    for start, end in song_bouts:
        ax2.axvspan(start/1000, end/1000, alpha=0.15, color='green')

    # Add summary text
    summary_text = f"Detection: {len(onsets)} syllables, {len(song_bouts)} song bouts"
    if len(onsets) > 0:
        durations = offsets - onsets
        summary_text += f"\nSyllable durations: {durations.mean():.1f}±{durations.std():.1f}ms"

    fig.text(0.02, 0.02, summary_text, fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgray"))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.15)

    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        print(f"Plot saved to: {save_path}")
    else:
        plt.savefig('amplitude_detection.png', dpi=150, bbox_inches='tight')
        print("Plot saved to: amplitude_detection.png")

    plt.close()

    return onsets, offsets, song_bouts

# Test visualization with a good threshold
print("Creating visualization...")
onsets, offsets, song_bouts = visualize_amplitude_detection(
    converted_test_file, threshold_percentile=80
)

print(f"\nResults: {len(onsets)} syllables, {len(song_bouts)} song bouts")
```

```python
def visualize_amplitude_detection_with_spectrogram(audio_file, threshold_percentile=80, save_path=None):
    """Visualize amplitude thresholding detection with spectrogram"""
    from tools.audio_utils import read_audio_file
    from scipy.signal import spectrogram

    audio, fs = read_audio_file(audio_file)

    # Get smoothed signal and detections
    smooth = smooth_data_matlab_style(audio, fs)
    onsets, offsets = detect_syllables_amplitude_threshold(
        audio, fs, threshold_percentile=threshold_percentile
    )
    song_bouts = find_song_bouts(onsets, offsets)

    # Calculate threshold value
    threshold_value = np.percentile(smooth, threshold_percentile)

    # Create spectrogram
    nperseg = 1024
    noverlap = int(nperseg * 0.8)
    f, t, Sxx = spectrogram(audio, fs, nperseg=nperseg, noverlap=noverlap)

    # Convert to dB and limit frequency range
    Sxx_db = 10 * np.log10(Sxx + 1e-10)
    freq_mask = (f >= 500) & (f <= 10000)  # Focus on birdsong frequencies
    f_plot = f[freq_mask]
    Sxx_plot = Sxx_db[freq_mask, :]

    # Time axes
    audio_time = np.arange(len(audio)) / fs
    smooth_time = np.arange(len(smooth)) / fs

    # Create figure with 3 subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))

    # 1. Raw audio with detections
    ax1.plot(audio_time, audio, 'b-', alpha=0.7)
    ax1.set_ylabel('Amplitude')
    ax1.set_title(f'Raw Audio: {os.path.basename(audio_file)}')
    ax1.grid(True, alpha=0.3)

    # Mark syllables and song bouts on audio
    for onset, offset in zip(onsets/1000, offsets/1000):
        ax1.axvspan(onset, offset, alpha=0.3, color='red',
                   label='Syllables' if onset == onsets[0]/1000 else "")

    for start, end in song_bouts:
        ax1.axvspan(start/1000, end/1000, alpha=0.15, color='green',
                   label='Song Bouts' if start == song_bouts[0][0] else "")

    if len(onsets) > 0:
        ax1.legend()

    # 2. Smoothed signal with threshold
    ax2.plot(smooth_time, smooth, 'k-', linewidth=1, label='Smoothed Signal')
    ax2.axhline(y=threshold_value, color='red', linestyle='--',
               label=f'{threshold_percentile}th percentile: {threshold_value:.6f}')
    ax2.set_ylabel('Smoothed Power')
    ax2.set_title('Smoothed Signal with Amplitude Threshold')
    ax2.grid(True, alpha=0.3)
    ax2.legend()

    # Mark detections on smoothed signal
    for onset, offset in zip(onsets/1000, offsets/1000):
        ax2.axvspan(onset, offset, alpha=0.3, color='red')

    for start, end in song_bouts:
        ax2.axvspan(start/1000, end/1000, alpha=0.15, color='green')

    # 3. Spectrogram with detections
    im = ax3.pcolormesh(t, f_plot/1000, Sxx_plot, shading='gouraud', cmap='inferno')
    ax3.set_ylabel('Frequency (kHz)')
    ax3.set_xlabel('Time (s)')
    ax3.set_title('Spectrogram with Detected Syllables')

    # Add colorbar
    cbar = plt.colorbar(im, ax=ax3)
    cbar.set_label('Power (dB)')

    # Mark syllables on spectrogram
    for onset, offset in zip(onsets/1000, offsets/1000):
        ax3.axvspan(onset, offset, alpha=0.4, color='white', linewidth=0)
        ax3.axvline(onset, color='white', linewidth=1, alpha=0.8)
        ax3.axvline(offset, color='white', linewidth=1, alpha=0.8)

    # Mark song bouts on spectrogram
    for start, end in song_bouts:
        ax3.axvspan(start/1000, end/1000, alpha=0.2, color='lime', linewidth=0)

    # Add summary text
    summary_text = f"Detection: {len(onsets)} syllables, {len(song_bouts)} song bouts"
    if len(onsets) > 0:
        durations = offsets - onsets
        summary_text += f"\nSyllable durations: {durations.mean():.1f}±{durations.std():.1f}ms"

    fig.text(0.02, 0.02, summary_text, fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgray"))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.12)

    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        print(f"Plot saved to: {save_path}")
    else:
        plt.savefig('amplitude_detection_with_spectrogram.png', dpi=150, bbox_inches='tight')
        print("Plot saved to: amplitude_detection_with_spectrogram.png")

    plt.close()

    return onsets, offsets, song_bouts

# Test visualization with spectrogram
print("Creating visualization with spectrogram...")
onsets, offsets, song_bouts = visualize_amplitude_detection_with_spectrogram(
    converted_test_file, threshold_percentile=80
)

print(f"\nResults: {len(onsets)} syllables, {len(song_bouts)} song bouts")
```
