```python
import os

def simple_drive_test():
    """Simple hello world file test on E: drive"""

    print("Testing E: drive with simple file operations...")

    # Test file path
    test_file = r'E:\xfosters\bk1bk3\hello_test.txt'

    try:
        # Test 1: Write file
        print("1. Writing hello world file...")
        with open(test_file, 'w') as f:
            f.write("Hello World!")
        print(f"   ✓ Successfully wrote to: {test_file}")

        # Test 2: Read file
        print("2. Reading hello world file...")
        with open(test_file, 'r') as f:
            content = f.read()
        print(f"   ✓ Successfully read: '{content}'")

        # Test 3: Clean up
        print("3. Deleting test file...")
        os.remove(test_file)
        print("   ✓ File deleted successfully")

        print("\n✅ E: drive read/write works fine!")

    except Exception as e:
        print(f"\n❌ Error: {e}")
        print(f"Error type: {type(e).__name__}")

# Run the test
simple_drive_test()
```

```python
import os

def debug_drive():
    """Debug E: drive issues"""

    print("Debugging E: drive...")

    # Test 1: Check if E: exists at all
    print(f"E: exists: {os.path.exists('E:')}")
    e_root = 'E:\\'
    print(f"E:\\ exists: {os.path.exists(e_root)}")

    # Test 2: Try to list E: contents
    try:
        contents = os.listdir('E:\\')
        print(f"E:\\ contents: {contents[:5]}...")  # First 5 items
    except Exception as e:
        print(f"Cannot list E:\\: {e}")

    # Test 3: Try writing to E: root
    try:
        with open('E:\\test.txt', 'w') as f:
            f.write("test")
        print("✓ Can write to E: root")
        os.remove('E:\\test.txt')
    except Exception as e:
        print(f"Cannot write to E: root: {e}")

    # Test 4: Check the specific xfosters path
    xfosters_path = 'E:\\xfosters'
    print(f"E:\\xfosters exists: {os.path.exists(xfosters_path)}")

    # Test 5: Try different path formats
    test_paths = [
        'E:/xfosters/bk1bk3/test.txt',  # Forward slashes
        'E:\\xfosters\\bk1bk3\\test.txt',  # Backslashes
        r'E:\xfosters\bk1bk3\test.txt',  # Raw string
    ]

    for path in test_paths:
        try:
            with open(path, 'w') as f:
                f.write("test")
            print(f"✓ Success with: {path}")
            os.remove(path)
            break
        except Exception as e:
            print(f"✗ Failed with {path}: {e}")

debug_drive()
```
